```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('API_SP.POP.GROW_DS2_en_excel_v2_4770572.xls')
df = df.iloc[3:]
df
```

```python
#Renaming Unnamed Columns as years from 1961 to 2021
# create a list of old column names
old_cols = [f'Unnamed: {i}' for i in range(5,65)]
# create a list of new column names
new_cols = [str(i) for i in range(1961,2021)]
# create a dictionary to map old column names to new column names
col_map = dict(zip(old_cols, new_cols))
# use the rename function to map old column names to new column names
df = df.rename(columns=col_map)
df = df.rename(columns={'Unnamed: 65':'2021'})

df.head()
```

```python
columns_to_remove = ["World Development Indicators","Unnamed: 2", "Unnamed: 3", "Unnamed: 4"]
df = df.drop(columns_to_remove, axis=1)

#Rename Data Source Column to Country for better understanding of dataset
df = df.rename(columns={'Data Source': 'Country'})
```

Defining Functions to draw plots

```python


def plot_decade_population_growth(df, countries):
    '''
    Plots a line chart of the average population growth rate by decade for selected countries.

    Parameters:
    df (pandas.DataFrame): the dataframe containing the population growth rate data
    countries (list): a list of country names (strings) to include in the plot

    Returns:
    None
    '''
    # Filter the data to only include selected countries
    df_for_line_plot = df[df['Country'].isin(countries)]

    # Extract the columns for years 1961-2021
    columns = [str(year) for year in range(1961, 2022)]
    df_for_line_plot = df_for_line_plot[['Country'] + columns]

    # Set the country column as the index
    df_for_line_plot = df_for_line_plot.set_index('Country')

    # Transpose the DataFrame so that each row represents a year
    df_for_line_plot = df_for_line_plot.transpose()

    # Convert the index to a DateTimeIndex
    df_for_line_plot.index = pd.to_datetime(df_for_line_plot.index)

    # Resample the data by decade and calculate the average population growth rate
    decade_growth_rate = df_for_line_plot.resample('10Y').mean().pct_change()

    # Increase the width of the plot
    plt.figure(figsize=(10, 6))

    # Plot the average population growth rate as a line chart
    plt.plot(decade_growth_rate.index.year, decade_growth_rate)

    # Add labels and a legend to the chart
    plt.title('Average Population Growth Rate by Decade')
    plt.xlabel('Decade')
    plt.ylabel('Population Growth Rate (%)')
    plt.legend(countries)

    # Display the chart
    plt.show()

```

```python


def plot_box_region_pop_growth(df, year):
    """
    Plot a box plot of population growth rate by region for a specific year.
    
    Parameters:
    -----------
    df : pandas dataframe
        A dataframe containing population growth rate data by country and year.
    year : int
        The year to plot the box plot for.
    
    Returns:
    --------
    None
    """
    # Create a new column for the region
    region_dict = {'North America': ['Canada', 'United States', 'Mexico'],
               'South America': ['Brazil', 'Argentina', 'Peru'],
               'Europe': ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain'],
               'Asia': ['China', 'India', 'Japan', 'Indonesia', 'Philippines'],
               'Africa': ['Nigeria', 'Egypt', 'South Africa', 'Ethiopia', 'Kenya']}

    region_list = []
    for country in df['Country']:
        for region, countries in region_dict.items():
            if country in countries:
                region_list.append(region)
                break
        else:
            region_list.append('Other')
    df['Region'] = region_list

    # Create the box plot
    plt.figure(figsize=(10,6))
    plt.boxplot([df[df['Region']=='North America'][str(year)],
                 df[df['Region']=='South America'][str(year)],
                 df[df['Region']=='Europe'][str(year)],
                 df[df['Region']=='Asia'][str(year)],
                 df[df['Region']=='Africa'][str(year)]],
                labels=['North America', 'South America', 'Europe', 'Asia', 'Africa'])
    plt.title(f'Population Growth Rate by Region ({year})')
    plt.ylabel('Annual Population Growth Rate (%)')
    plt.show()

```

```python


def top_10_countries_by_population_Growth_rate_in_2021(df: pd.DataFrame):
    """
    Creates a bar chart of the top 10 most populous countries in 2021.

    Parameters:
    df (pd.DataFrame): A pandas DataFrame containing population data for multiple countries and years.

    Returns:
    None
    """
    # Filter the data for the year 2021
    df_2021 = df[['Country', '2021']]

    # Sort the data by population growth in descending order
    df_2021_sorted = df_2021.sort_values(by='2021', ascending=False)

    # Take the top 10 countries by population
    df_2021_top10 = df_2021_sorted.head(10)

    # Create a bar chart
    plt.figure(figsize=(12,8))
    plt.bar(df_2021_top10['Country'], df_2021_top10['2021'])
    plt.title('Top 10 Countries of the most population growth in 2021')
    plt.xlabel('Country')
    plt.ylabel('Population Growth Rate')
    plt.xticks(rotation=45, ha='right')
    plt.show()

```
